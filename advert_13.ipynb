# KNN

**previewing the target variable**

```python
%%R
table(df$Clicked.on.Ad)
```

```python
%%R
# Storing it as data frame
adds <- data.frame(df)
```

**Randomnization**

```python
%%R
# Creating a random number equal 90% of total number of rows
ran <- sample(1:nrow(adds),0.9 * nrow(adds))
```

**Normalization**

```python
%%R
# The normalization function is created
nor <-function(x) { (x -min(x))/(max(x)-min(x))   }
```

```python
%%R
# Normalization function is applied to the dataframe
dia_nor <- as.data.frame(lapply(adds[,c(1,2,3,4)], nor))
 
```

**Extracting training dataset**

```python
%%R
# The training dataset extracted
dia_train <- dia_nor[ran,]
head(dia_train)
```

**Extracting test dataset**

```python
%%R
# The test dataset extracted
dia_test <- dia_nor[-ran,]
tail(dia_test)
```

**creating target df**

```python
%%R
# The 10th column of training dataset because that is what we need to predict about testing dataset
# also convert ordered factor to normal factor
dia_target <- as.factor(adds[ran,10])
```

```python
%%R
# The actual values of 10th coulmn of testing dataset to compaire it with values that will be predicted
# also convert ordered factor to normal factor
test_target <- as.factor(adds[-ran,10])
test_target
```

**Running KNN function**

```python
%%R
# Running the knn function
library(class)
#using k nearest neighbours of 20
pr <- knn(dia_train,dia_test,cl=dia_target,k=20)
 
# Creating the confucion matrix
tb <- table(pr,test_target)
tb
 
# Checking the accuracy
accuracy <- function(x){sum(diag(x)/(sum(rowSums(x)))) * 100}
accuracy(tb)
```

**Decresing the number of k to see what happens to the accuracy**

```python
%%R
# Running the knn function
library(class)
pr <- knn(dia_train,dia_test,cl=dia_target,k=5)
 
# Creating the confucion matrix
tb <- table(pr,test_target)
tb
 
# Checking the accuracy
accuracy <- function(x){sum(diag(x)/(sum(rowSums(x)))) * 100}
accuracy(tb)
```

**splitting data into train and test**

```python
%%R
# Next i split the data into training set and testing set. 
# NB: The training set will be used for model building while the testing set for model evaluation.
# ---
# - The “y” parameter takes the value of variable according to which data needs to be partitioned. 
# In my case, target variable is at clicked.on.add

# We are using p=0.7. It means that data split should be done in 70:30 ratio. 
# So, 70% of the data is used for training and the remaining 30% is for testing the model.

intrain <- createDataPartition(y = df$Clicked.on.Ad, p= 0.7, list = FALSE)
training <- df[intrain,]
testing <- df[-intrain,]
```

```python
%%R
install.packages("e1071", dep = TRUE) 
```

```python
%%R
library(e1071)
```

```python
%%R
training[["Clicked.on.Ad"]] = factor(training[["Clicked.on.Ad"]])
```

```python
%%R
# Before i train my model i will need to control all the computational overheads. 
# i will implement this through the trainControl() method. 
# This will allow me to use the train() function provided by the caret package. 
# ---
# The trainControl method will take three parameters:
# a) The “method” parameter defines the resampling method, 
# in this demo we’ll be using the repeatedcv or the repeated cross-validation method.
# b) The next parameter is the “number”, this basically holds the number of resampling iterations.
# c) The “repeats ” parameter contains the sets to compute for our repeated cross-validation. 
# i am  using setting number =10 and repeats =3
# ---
# 
trctrl <- trainControl(method = "repeatedcv", number = 10, repeats = 3)

svm_Linear <- train(Clicked.on.Ad ~., data = training, method = "svmLinear",
trControl=trctrl,
preProcess = c("center", "scale"),
tuneLength = 10)
```

**model performance**

```python

%%R
# i can then check the reult of my train() model as shown below
# ---
# 
svm_Linear
```

My model performed quite well with  an accuracy of 93%

factorizing

```python
#factorise
%%R
df$Clicked.on.Ad <- factor(df$Clicked.on.Ad, levels = c(0,1), labels = c("False", "True"))
```

**splitting data into training and test**

```python
%%R
# Splitting data into training and test data sets
# ---
# 
indxTrain <- createDataPartition(y = df$Clicked.on.Ad,p = 0.75,list = FALSE)
training <- df[indxTrain,]
testing <- df[-indxTrain,]
 
```

**Checking dimensions of the split**

```python
%%R
# Checking dimensions of the split
# ---
#
prop.table(table(df$Clicked.on.Ad)) * 100
prop.table(table(training$Clicked.on.Ad)) * 100
prop.table(table(testing$Clicked.on.Ad)) * 100
```

```python
%%R
# Comparing the outcome of the training and testing phase
# ---
# Creating objects x which holds the predictor variables and y which holds the response variables
# ---
#
x =  training[,-10]
 
y = training$Clicked.on.Ad
```

**building model**

```python
%%R
# Now building my model 
# ---
# 
model = train(x,y,'nb',trControl=trainControl(method='cv',number=10))
```

**Model Evaluation**

```python
%%R
# Model Evalution
# ---
# Predicting my testing set
# 
Predict <- predict(model,newdata = testing )
```

**Confusion Matrix**

```python
%%R
lll <- confusionMatrix(Predict, testing$Clicked.on.Ad)  
lll
```

model performed quite well with an accuracy of 95%
